```python
# Import our dependencies
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
import tensorflow as tf

#  Import and read the diabetes_data.csv.
import pandas as pd
diabetes_df = pd.read_csv('https://raw.githubusercontent.com/ComfyKoala/diabetes-classification/main/Maribel/diabetes_data_upload.csv')
# Display the first few rows
diabetes_df.head()
```

```python
#Dropping weakest features "muscle stiffness", "Obesity", "weakness","sudden weight loss","partial paresis"
diabetes_df = diabetes_df.drop(columns=["muscle stiffness", "Obesity", "weakness","sudden weight loss","partial paresis"])
diabetes_df.head(5)
```

```python
#change all "yes" to 1 and all "no" to 0
diabetes_df = diabetes_df.replace({"Yes": 1, "No": 0})

#change "female" to 1 and "male" to 0
diabetes_df = diabetes_df.replace({"Female": 1, "Male": 0})
```

```python
# Get the target variables
target=diabetes_df["class"]
target_names = ["negative", "positive"]
```

```python
#Get the features
data = diabetes_df.drop("class", axis=1)
feature_names = data.columns
data.head()
```

```python
# Split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

## Support Vector Machines

```python
# Support vector machine linear classifier
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
# Model Accuracy
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
# Calculate the classification report
from sklearn.metrics import classification_report
predictions = model.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```

Removed weakest columns and accuracy went from 92.3% in file of "diabetes1updated" to 93.8% above

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier
```

```python
# Train the data
classifier.fit(X_train, y_train)
```

```python
# Score the model
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
# Predict outcomes for test data set
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
from sklearn.metrics import accuracy_score
# Display the accuracy score for the test dataset.
accuracy_score(y_test, predictions)
```
